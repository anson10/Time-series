# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
register_matplotlib_converters()
```

# Data Loading

```python
df = pd.read_csv('../data/icecream.csv')
```

```python
df.head()
```

```python
# columns rename
df.rename(columns={'DATE': 'date', 'IPN31152N': 'prod'}, inplace=True)
```

```python
# date -> index
df.set_index('date', inplace=True)
```

```python
# data subsetting from 2010
start_date = pd.to_datetime('2010-01-01')
df.index = pd.to_datetime(df.index)
df = df[start_date:]
```

```python
df.head()
```

```python
# plotting the data
plt.figure(figsize=(12, 6))
plt.plot(df['prod'])
plt.title('Ice Cream Production over Time', fontsize=20)
plt.ylabel('Production Index', fontsize=16)
for year in range(2010, 2021):
    plt.axvline(pd.to_datetime(str(year)+'-01-01'), color='k', linestyle='--', linewidth=0.5)
plt.show()
```

# ACF Plot

```python
acf_plot = plot_acf(df['prod'],lags=100)
```

# PACF Plot

```python
plot_plot = plot_pacf(df['prod'])
```
